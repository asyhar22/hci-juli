```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/asyhar22/hci-juli/main/Marketing_Campaign_Data_A.csv'
df_campaign = pd.read_csv(url)
df_campaign                   
```

```python
df_campaign_copy = df_campaign.copy()
df_campaign_copy
```

```python
from datetime import date, datetime
```

```python
df_campaign_copy['APPL_DATETIME'] = pd.to_datetime(df_campaign_copy['APPL_DATETIME'])
df_campaign_copy['APPL_DATETIME']
```

```python
df_campaign_copy['DATE_BIRTH'] = pd.to_datetime(df_campaign_copy['DATE_BIRTH'])
df_campaign_copy['DATE_BIRTH']
```

```python
df_campaign_copy['AGE'] = (df_campaign_copy['APPL_DATETIME'] - df_campaign_copy['DATE_BIRTH']) / np.timedelta64(1,'Y')
df_campaign_copy['AGE']
```

```python
df_campaign_copy['AGE CATEGORY'] = ['Late_Ten' if x < 20 else 'Twenty' if x < 30 else 'Thirty' if x < 40 else 'Fourty' if x < 50 else 'Fifty' if x < 60 else 'Above_Sixty' if x >= 60 else 'Uncategorized' for x in df_campaign_copy['AGE']]
df_campaign_copy['AGE CATEGORY']
```

```python
df_campaign_copy['AGE CATEGORY'].unique()
```

```python
df_campaign_copy.groupby(['AGE CATEGORY']).count()['TARGET'].sort_values(ascending=False)
```
